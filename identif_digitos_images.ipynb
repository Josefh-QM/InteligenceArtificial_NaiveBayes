```python
# digitos para reconocimiento ia

import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# Cargar el dataset de dígitos
digits = load_digits()
X = digits.data
y = digits.target

# Dividir el dataset en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# Crear el modelo de MLPClassifier
mlp_digits = MLPClassifier(hidden_layer_sizes=(50, 30), max_iter=1000, random_state=42)

```

```python
# Entrenar el modelo
mlp_digits.fit(X_train, y_train)

```

```python
# Hacer predicciones
y_pred_train = mlp_digits.predict(X_train)
y_pred_test = mlp_digits.predict(X_test)

# Evaluar el modelo
accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print(f"Exactitud en el conjunto de entrenamiento: {accuracy_train:.2f}")
print(f"Exactitud en el conjunto de prueba: {accuracy_test:.2f}")

# Mostrar el reporte de clasificación
print("Reporte de Clasificación en el Conjunto de Prueba:")
print(classification_report(y_test, y_pred_test))

# Graficar algunas imágenes de dígitos con sus predicciones
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for ax, image, label in zip(axes.flatten(), X_test, y_pred_test):
    ax.imshow(image.reshape(8, 8), cmap='gray')
    ax.set_title(f'Pred: {label}')
    ax.axis('off')
plt.show()
```

```python
train_sizes, train_scores, test_scores = learning_curve(
    mlp_digits, X_train, y_train, cv=5, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10), random_state=42)

# Calcular medias y desviaciones estándar
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
```

```python
# Graficar la curva de aprendizaje
plt.figure()
plt.title("Curva de aprendizaje")
plt.xlabel("Tamaño del conjunto de entrenamiento")
plt.ylabel("Puntuación")
plt.grid()

# Área para desviaciones estándar
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1, color="r")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.1, color="g")

# Graficar medias
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Puntuación de entrenamiento")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Puntuación de validación cruzada")

plt.legend(loc="best")
plt.show()
```
